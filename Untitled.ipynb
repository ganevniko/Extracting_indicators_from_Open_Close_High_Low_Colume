```python
from local_Functions import RSI, targetColumn
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('timeSeries.csv')
df=df.drop(columns='Adj Close',axis=1)
df=df.set_index('Date')
tempColumns=list(df.columns)
df['targetColumn']=targetColumn(df.Close,5)

tailSize, bodyRange=[], []

for i in df.index: 
    #Green Candle case
    if df.Close[i]-df.Open[i]>0: 
        t=(df.High[i]-df.Close[i])/(df.High[i]-df.Low[i])
        tailSize.append(t)
        bodyRange.append(df.Close[i]-df.Open[i])
    #Red Candle Case
    else: 
        t=(df.Close[i]-df.Low[i])/(df.High[i]-df.Low[i])
        tailSize.append(t)
        bodyRange.append(df.Open[i]-df.Close[i])
df["tailSize"]=tailSize
df["bodyRange"]=bodyRange
tempColumns.append("bodyRange")



df["RSI"]=RSI(df.Close)
df["RSI"]=df['RSI']/100
df["totalRange"]=(df.High-df.Low)
df["closeChange"]=df.Close.pct_change()
tempColumns.append("totalRange")
tempColumns.append("closeChange")
df["openChange"]=df.Open.pct_change()
df["highChange"]=df.High.pct_change()
df["lowChange"]=df.Low.pct_change()
df["volumeChange"]=df.Volume.pct_change()
df["rsiChange"]=df.RSI.pct_change()
df["tailSizeChange"]=df.tailSize.pct_change()
df["bodyRangeChange"]=df.bodyRange.pct_change()
df["totalRangeChange"]=df.totalRange.pct_change()
tempColumns.append("bodyRangeChange")
tempColumns.append("totalRangeChange")


df["volumeChange_mva"]=df.volumeChange.rolling(20).mean()
df["tailSizeChange_mva"]=df.tailSizeChange.rolling(20).mean()

df["bodyRangeChange_std"]=df.bodyRangeChange.rolling(20).std()
df["totalRangeChange_std"]=df.totalRangeChange.rolling(20).std()

df["bodyRangeChange_vs_std"]=df.bodyRangeChange/df.bodyRangeChange_std
df["totalRangeChange_vs_std"]=df.totalRangeChange/df.totalRangeChange_std
df["volumeChange_vs_mva"]=df.volumeChange/df.volumeChange_mva




```
